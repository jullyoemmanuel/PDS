## 2.1 DFT e amostragem

Os sinais analisados são de um mesmo tipo, e apenas seus parâmetros variam. Teremos uma função cosseno na forma $x(t) = cos(w_0 t)$, que será amostrada com uma taxa $f_s$, utilizando uma quantidade $N$ de pontos. Para facilitar a tarefa, geramos uma função que recebe em suas entradas os parâmetros $\frac{w_0}{\pi}$ (que será igual a 2 ou 3 para cada letra da questão), $f_s$ e $N$, e então gera $x(t)$, $X(jw)$ e a fase de $X(jw)$ com seus respectivos gráficos.

```python
# Função que gera x(t) X(jw), Xp(jw) e xp(t)
def cosine(w0,fs,N):
  Ts = 1 / fs # Período de amostragem
  Tw = (N-1)*Ts # Tamanho da janela
  t = np.arange(0,Tw,Ts) # Gera o vetor tempo do sinal
  xt = np.cos(w0* np.pi * t) # Gera o sinal
  plt.plot(t,xt) # plota o sinal
  plt.show()
  # FFT
  fw = 1 / Tw # Passo na frequência
  w = np.arange((-N/2)*fw,(N/2-1)*fw,fw) # Gera o range de frequências
  FT = np.fft.fftshift(np.fft.fft(xt)) # Gera a FFT
  plt.plot(w,np.absolute(FT)) # Plota
  plt.show()
  # Fase da FFT
  plt.plot(np.angle(FT))
  plt.show()
  return xt,FT
```

#### $x(t) = cos(2\pi t), f_s=2, N=16$

```python
a1,b1 = cosine(2,2,16) # Executando cada etapa da função e salvando o vetor de x(t) e de X(jw) nas variáveis a1 e b1
```
