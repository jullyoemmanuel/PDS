# 1.1 Definição da TZ Bilateral

A Transformada Z Bilaterial é representada da seguinte forma:
$ X(z) = \sum_{n=-\infty }^{+\infty} x(n)z^{-n}$

----------------------------------------

1.1.1  $x(n) = u[n] - u[-n]$

Pela propriedade da linearidade, iremos calcular $X_{1}(z)$ e $X_{2}(z)$ separadamente e depois subtraí-los.

$X_{1}(z) = \frac{z}{z-1}$, para $|z| > 1$

Para calcular $X_{2}(z)$, iremos usar a propriedade de inversão no tempo, logo:

$X_{2}(z) = \frac{z-1}{z} $, para  $|z| > 1$

Por fim, 

$X(z) = \frac{z}{z-1} - \frac{z-1}{z} $, para $|z| > 1.$

-------------------------------------------

1.1.2 $x(n) = u[n] - u[n-10]$

Da mesma forma que a questão anterior, temos que:

$X_{1}(z) = \frac{z}{z-1}$, para $|z| > 1$

E, usando a propriedade da translação no tempo:

$X_{2}(z) = z^{-10}\frac{z}{z-1} $, para $|z| > 1\pm0.$ 

Por fim, 

$X(z) = \frac{z}{z-1} - z^{-10}\frac{z}{z-1}$ para $|z| > 1$.

# 1.4 TZ Inversa


1.4.1 $X(Z)=\frac{Z}{Z-\frac{1}{2}}$: 

Sabendo que $x[n]=a^n\mu[n] \longleftrightarrow X(Z)=\frac{Z}{Z-\alpha} \therefore \frac{Z}{Z-\frac{1}{2}} \longleftrightarrow (\frac{1}{2})^n\mu[n]$

------
1.4.2 $X(Z)=\frac{Z-1}{Z^2-2Z+1}=\frac{Z}{(Z-1)^2}-\frac{1}{(Z-1)^2}$

Sabendo que $x[n]=n\mu[n] \longleftrightarrow X(Z)=\frac{Z}{(Z-1)^2}$ e que $\frac{1}{(Z-1)^2}=Z^{-1}\frac{Z}{(Z-1)^2}$ 

E utilizando a propriedade do deslocamento no tempo:

$x[n] \longleftrightarrow X(Z)$

$x[n-n_0] \longleftrightarrow Z^{-n_0}X(Z)$

Obtemos $\frac{Z}{(Z-1)^2}-\frac{1}{(Z-1)^2} \longleftrightarrow n\mu[n]-(n-1)\mu[n-1]$

-------
1.4.3 $X(Z)=\frac{1}{Z(Z-\frac{1}{2})}$:

Para resolver esta questão, podemos usar a propriedade de deslocamento no tempo.
Assim, teremos $X'(Z)=\frac{1}{Z-\frac{1}{2}} \longleftrightarrow x'[n]=(\frac{1}{2})^n\mu[n] \therefore X(Z)=Z^{-1}\frac{1}{Z-\frac{1}{2}} \longleftrightarrow x[n]=x'[n-1]=(\frac{1}{2})^{n-1}\mu[n-1]=2(\frac{1}{2})^n\mu[n]$

------
1.4.4 $X(Z)=\frac{Z^2}{(Z-\frac{1}{2})(Z-\frac{3}{4})}$

Utilizando expansão em frações parciais, teremos $X'(Z)=\frac{Z}{(Z-\frac{1}{2})(Z-\frac{3}{4})}= \frac{A}{Z-\frac{1}{2}}+\frac{B}{Z-\frac{3}{4}}=\frac{(A+B)Z+(-\frac{3}{4}A-\frac{1}{2}B)}{(Z-\frac{1}{2})(Z-\frac{3}{4})} \therefore A+B=1$ e $-\frac{3}{4}A-\frac{1}{2}B=0 \therefore A=-2$ e $B=3$

Assim, utilizando a propriedade da linearidade, $X'(Z)=\frac{Z}{(Z-\frac{1}{2})(Z-\frac{3}{4})}= -\frac{2}{Z-\frac{1}{2}}+\frac{3}{Z-\frac{3}{4}} \longleftrightarrow x'[n]=-2(\frac{1}{2})^n\mu[n]+3(\frac{3}{4})^n\mu[n]$

Note que $X(Z)=\frac{Z^2}{(Z-\frac{1}{2})(Z-\frac{3}{4})}=Z^{+1}\frac{Z}{(Z-\frac{1}{2})(Z-\frac{3}{4})}=Z^{+1}X'(Z)$, e portanto podemos utilizar a propriedade de deslocamento no tempo para obter $x[n]=x'[n+1]=-2(\frac{1}{2})^{n+1}\mu[{n+1}]+3(\frac{3}{4})^{n+1}\mu[{n+1}]=-(\frac{1}{2})^{n}\mu[{n+1}]+\frac{9}{4}(\frac{3}{4})^{n}\mu[{n+1}]$

$2y[n+1]-y[n]=sin(\frac{\pi n}{16})u[n]$, e condição inicial: $y[0]=1$

---
Aplicando a transformada Z em ambos os lados:

$2z(Y(z)-y[0])-Y(z)=\frac{zsin(\frac{\pi}{8})}{z^{2}-2zcos(\frac{\pi}{8})+1}$

Substituindo $y[0]=1$:

$2zY(z)-2z-Y(z)=\frac{zsin(\frac{\pi}{8})}{z^{2}-2zcos(\frac{\pi}{8})+1}$

Isolando $Y(z)$:

$Y(z)=\frac{zsin(\frac{\pi}{8})}{(2z-1)(z^{2}-2zcos(\frac{\pi}{8})+1)}+\frac{2z}{2z-1}$

$Y(z)=\frac{zsin(\frac{\pi}{8})}{(z-\frac{1}{2})(z^{2}-2zcos(\frac{\pi}{8})+1)}+\frac{z}{z-\frac{1}{2}}$


---


Usando frações parciais:

$Y(z)=\frac{\frac{\frac{1}{4}sin(\frac{\pi}{8})}{\frac{5}{4}-cos(\frac{\pi}{8})}}{(z-\frac{1}{2})}+\frac{Az+B}{(z^{2}-2zcos(\frac{\pi}{8})+1)}+\frac{z}{z-\frac{1}{2}}$ 

$Y(z)=\frac{0.293}{(z-\frac{1}{2})}+\frac{Az+B}{(z^{2}-2zcos(\frac{\pi}{8})+1)}+\frac{z}{z-\frac{1}{2}}$

Se $z=0$ em:

$\frac{zsin(\frac{\pi}{8})}{(2z-1)(z^{2}-2zcos(\frac{\pi}{8})+1)}=\frac{0.293}{(z-\frac{1}{2})}+\frac{Az+B}{(z^{2}-2zcos(\frac{\pi}{8})+1)}$

Então $B=0.586$

Usando $z=1$ em:

$\frac{zsin(\frac{\pi}{8})}{(2z-1)(z^{2}-2zcos(\frac{\pi}{8})+1)}=\frac{0.293}{(z-\frac{1}{2})}+\frac{Az+0.586}{(z^{2}-2zcos(\frac{\pi}{8})+1)}$

Produz $A=-0.293$

Substituindo A e B na equação:

$Y(z)=\frac{0.293}{(z-\frac{1}{2})}+\frac{-0.293z+0.586}{(z^{2}-2zcos(\frac{\pi}{8})+1)}+\frac{z}{z-\frac{1}{2}}$

$Y(z)=\frac{0.293}{(z-0.5)}+\frac{-0.293z+0.586}{(z^{2}-1.847z+1)}+\frac{z}{z-0.5}$


---


Da tabela de transformadas, $\mathcal{Z}[sin(\omega_{0}n)u[n]]=\frac{zsin(\omega_0)}{z^{2}-2zcos(\omega_{0})+1}$, se $|z|>1$

e

$\mathcal{Z}[cos(\omega_{0}n)u[n]]=\frac{z[z-cos(\omega_0)]}{z^{2}-2zcos(\omega_{0})+1}$, se $|z|>1$

Da equação encontramos $\omega_{0}=0.392$

$Y(z)=\frac{0.293}{(z-0.5)}+\frac{z}{z-0.5}-0.293z^{-1}(\frac{z^{2}-0.923z}{z^{2}-1.847z+1}+\frac{-2z+0.923z}{z^{2}-1.847z+1})$

$Y(z)=\frac{0.293}{(z-0.5)}+\frac{z}{z-0.5}-0.293z^{-1}(\frac{z^{2}-0.923z}{z^{2}-1.847z+1}-2.812\frac{0.382z}{z^{2}-1.847z+1})$


---


Aplicando a transformada Z inversa, então:

$y[n]=0.293(\frac{1}{2})^{n-1}u[n-1]+(\frac{1}{2})^{n}u[n]-0.293[cos(\frac{\pi}{8}(n-1))-2.812sin(\frac{\pi}{8}(n-1))]u[n-1]$

```python
n = np.arange(0,41)
y = (0.293*(1/2)**(n-1))*(np.heaviside(n-1,1))+((1/2)**(n))*(np.heaviside(n,1))-0.293*(np.cos(((np.pi)/8)*(n-1))-2.812*(np.sin(((np.pi)/8)*(n-1)))*(np.heaviside(n-1,1)))

plt.stem(n,y,use_line_collection= True)
plt.xlabel('n')
plt.ylabel(' y[n]')
plt.show()
```

## 1.6 Estabilidade de Sistemas de Tempo Discreto

$
1.6.1 H(Z) = \frac{z}{z-\frac{2}{5}}$

Primeiro, para analisar a estabilidade, iremos olhar os polos da mesmas, neste caso temos apenas um polo, cujo módulo é menor que 1, logo, o sistema é estável.

```python
w = np.linspace(-10,10)
H = (np.exp(w*1j))/(np.exp(w*1j) - 0.4)

plt.stem(w,abs(H))
plt.xlabel('w')
plt.ylabel(' H[w]')
plt.show()
```

$ 1.6.2  H(z) = \frac{z}{z^2 - \frac{4}{25}}$

Ao fazermos algumas operações básicas com o denominador, encontramos:

$ z^2 - \frac{4}{25} = (z + \frac{2}{5})(z - \frac{2}{5})$

Logo, os polos serão - 0,4 e 0,4, em módulo, ambos são menores que 1, logo, o sistema é estável.

```python
w = np.linspace(-10,10)
H = (np.exp(w*1j))/((np.exp(w*1j))**2 - 0.24)

plt.stem(w,abs(H))
plt.xlabel('w')
plt.ylabel(' H[w]')
plt.show()
```

$1.6.3  H(z) = \frac{z}{z^3 - 8z^2 + 17z - 10}$

Ao fazermos algumas operações básicas com o denominador, encontramos:

$ z^3 - 8z^2 + 17z - 10 = (z - 1)(z + 2)(z + 5)$

Logo, os polos serão 1, -2 e -5, em módulo, ambos são maiores que 1, logo, o sistema é instável.

```python
w = np.linspace(-10,10)
H = (np.exp(w*1j))/((np.exp(w*1j))**3 - 8*((np.exp(w*1j))**2) + 17*(np.exp(w*1j)) - 10)

plt.stem(w,abs(H))
plt.xlabel('w')
plt.ylabel(' H[w]')
plt.show()
```

## 1.7 Estabilidade de Sistemas de Tempo Discreto

```python
K = 2
z = control.tf('z')
```

```python
polos = [control.pole(K / (1 + K * (z/ (z - 0.9))))[0] for K in np.arange(-100, 100, 0.01)]
```

```python
plt.plot(np.arange(-100, 100, 0.01), polos)
```

```python
valores = np.column_stack((list(np.arange(-100, 100, 0.01)), polos))
```

```python
print('Mínimo instável: ', valores[valores[:, 1] >0][:, 0].min())
print('Máximo instável: ', valores[valores[:, 1] >0][:, 0].max())
```

Dos valores calculados acima, infere-se que o sistema será estável para todos os valores de K exceto na faixa [-1, 100].
